### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each plot
x = np.linspace(0, 10, 100)
y_line = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot (first subplot)
axs[0, 0].plot(x, y_line)
axs[0, 0].set_title('Line Plot')

# Bar plot (second subplot)
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Plot')

# Scatter plot (third subplot)
axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')

# Pie chart (fourth subplot)
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

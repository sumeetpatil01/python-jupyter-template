### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['skyblue', 'salmon', 'lightgreen', 'gold']  # Different colors for each bar

# Create the bar plot
plt.bar(categories, values, color=colors)

# Add title
plt.title("Bar Plot of Categories and Values")

# Display the plot
plt.show()
```
